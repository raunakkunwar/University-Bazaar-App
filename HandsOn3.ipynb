```python
import time
import numpy as np
import matplotlib.pyplot as plt

def func(n):
    x = 1
    for i in range(n):
        for j in range(n):
            x = x + 1
    return x

def cal_time(n):
    start_time = time.time()
    func(n)
    end_time = time.time()
    return end_time - start_time

input_sizes = np.arange(1, 1001, 100)  
times = []

for n in input_sizes:
    times.append(cal_time(n))

# Fitting the data to a polynomial
p_coeff = np.polyfit(input_sizes, times, 2) 
fitted_poly = np.poly1d(p_coeff)

upper_bound = lambda n: 1.1 * (p_coeff[0] * n**2 + p_coeff[1] * n + p_coeff[2])

lower_bound = lambda n: 0.9 * (p_coeff[0] * n**2)

fine_input_sizes = np.linspace(1, 1000, 500)

plt.figure(figsize=(15, 8))
plt.plot(input_sizes, times, marker='o', label='Calculated Time')
plt.plot(fine_input_sizes, fitted_poly(fine_input_sizes), label='Fitted Polynomial', color='black')
plt.plot(fine_input_sizes, upper_bound(fine_input_sizes), label='Upper Bound', linestyle='--', color='orange')
plt.plot(fine_input_sizes, lower_bound(fine_input_sizes), label='Lower Bound', linestyle='--', color='red')

plt.xlabel('n (Input Size)')
plt.ylabel('Tn (Time Taken)')
plt.title('Tn vs n with Upper and lower bound')
plt.legend()
plt.show()
```

```python
import time
import numpy as np
import matplotlib.pyplot as plt

def func(n):
    x = 1
    for i in range(n):
        for j in range(n):
            x = x + 1
    return x

def cal_time(n):
    start_time = time.time()
    func(n)
    end_time = time.time()
    return end_time - start_time

input_sizes = np.arange(1, 1001, 100)  
times = []

for n in input_sizes:
    times.append(cal_time(n))

# Fitting the data to a polynomial
p_coeff = np.polyfit(input_sizes, times, 2) 
fitted_poly = np.poly1d(p_coeff)

upper_bound = lambda n: 1.1 * (p_coeff[0] * n**2 + p_coeff[1] * n + p_coeff[2])

lower_bound = lambda n: 0.9 * (p_coeff[0] * n**2)

fine_input_sizes = np.linspace(1, 1000, 500)

n_0 = 300  

plt.figure(figsize=(15, 8))
plt.plot(input_sizes, times, marker='o', label='Calculated Time')
plt.plot(fine_input_sizes, fitted_poly(fine_input_sizes), label='Fitted Polynomial', color='black')
plt.plot(fine_input_sizes, upper_bound(fine_input_sizes), label='Upper Bound', linestyle='--', color='orange')
plt.plot(fine_input_sizes, lower_bound(fine_input_sizes), label='Lower Bound', linestyle='--', color='red')

plt.axvline(x=n_0, color='blue', linestyle='--', label=f'n_0 = {n_0}')

plt.xlabel('n (Input Size)')
plt.ylabel('Tn (Time Taken)')
plt.title('Tn vs n with Upper and lower bound')
plt.legend()
plt.show()
```

## 5. Will this increate how long it takes the algorithm to run (e.x. you are timing the function like in #2)? 

Yes, new function takes little bit more time in compare to original function. but does not change the overall time complexity.(which is O(n^2))

```python
import time
import numpy as np
import matplotlib.pyplot as plt

def func(n):
    x = 1
    for i in range(n):
        for j in range(n):
            x = x + 1
    return x

def func_modify(n):
    x = 1
    y = 1
    for i in range(n):
        for j in range(n):
            x = x + 1
            y = i + j
    return x

def cal_time(n):
    start_time = time.time()
    func(n)
    end_time = time.time()
    time_func = end_time - start_time
    
    start_time = time.time()
    func_modify(n)
    end_time = time.time()
    time_func_modify = end_time - start_time
    
    return time_func, time_func_modify

input_sizes = np.arange(1, 1001, 100)  
times_func = []
times_func_modify = []

for n in input_sizes:
    time_func, time_func_modify = cal_time(n)
    times_func.append(time_func)
    times_func_modify.append(time_func_modify)

plt.figure(figsize=(15, 8))
plt.plot(input_sizes, times_func, marker='o', label='func Time', color='black')
plt.plot(input_sizes, times_func_modify, marker='o', label='func_modify Time', color='orange')

plt.xlabel('n (Input Size)')
plt.ylabel('Tn (Time Taken)')
plt.title('Tn vs n for func and func_modify')
plt.legend()
plt.show()
```

# 7. Merge sort

```python
def merge_sort(arr):
    if len(arr) > 1:
        midA = len(arr) // 2
        leftA = arr[:midA]
        rightA = arr[midA:]
        
        print(f"Splitting: {arr}")
        print(f"Left: {leftA}")
        print(f"Right: {rightA}")
        
        merge_sort(leftA)
        merge_sort(rightA)
        
        i = 0
        j=0
        k=0
        
        while i < len(leftA) and j < len(rightA):
            if leftA[i] < rightA[j]:
                arr[k] = leftA[i]
                i += 1
                k+=1
            else:
                arr[k] = rightA[j]
                j += 1
                k += 1
        
        while i < len(leftA):
            arr[k] = leftA[i]
            i += 1
            k += 1
        
        while j < len(rightA):
            arr[k] = rightA[j]
            j += 1
            k += 1

        print(f"Merging: {arr}")
    
    return arr

arr = [5,2,4,7,1,3,2,6]
print("Given array:", arr)
merge_sort(arr)
print("Final sorted array:", arr)
```
